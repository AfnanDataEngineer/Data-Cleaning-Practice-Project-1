```python
import pandas as pd
```

```python
df = pd.read_excel(r'D:\Data Engineering\Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex= True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' +x[6:10]).astype(str)
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

```python
df[["Street Address", "State","Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.fillna("")
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.strip()
df
```

```python
for x in df['Do_Not_Contact']:
    if df['Do_Not_Contact'] == "":
        df['Do_Not_Contact'].str.replace("","N")
        
```

```python
def replace_empty(cell):
    if cell == "":
        return "N"
    return cell

df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(replace_empty)
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
df.reset_index(drop=True)
```
